````python
import PyPDF2

def read_pdf_file (file_path : str) :

    text = """"""
    with open(file_path , 'rb') as file:
        pdf_reader = PyPDF2.PdfReader(file)

        for page_num in range(len(pdf_reader.pages)):
            page = pdf_reader.pages[page_num]
            text += page.extract_text()
    return text 

past_papers_data = read_pdf_file('./pdfs/June 2017 QP.pdf')
print(past_papers_data)
````

````python
from openai import OpenAI
from dotenv import load_dotenv, find_dotenv
import os

load_dotenv(find_dotenv())

client = OpenAI(api_key=os.getenv("OPENAI_API_KEY"))


def get_maked_down_questions (question : str , should_contain_diagram : bool) :
    response = client.chat.completions.create(
        model="gpt-4o-2024-08-06",
        messages=[
            {"role": "system", "content": SYSTEM_PROMPT},
            {
                "role": "user",
                "content": f"""
        Here is the question in raw form that needs to be transformed, the question is delimited in tripple backticks below:
        
        ```{{
                'question' : {question},
                'should_contain_diagram' : {should_contain_diagram}
            }}```
        """,
            },
        ],
        response_format={'type' : "json_object"},
    )

    return response.choices[0].message.content
````

````python
from pydantic import BaseModel, ValidationError
from uuid import uuid4
  

class JSON_QUESTION_SCHEMA (BaseModel):
    question : str 
    type : str 
    choices : list[str]
    answer : str
    diagram_description :str
    

class QUESTION_SCHEMA(BaseModel):
    id : int
    definition : str 
    should_contain_diagram : bool
    total_marks : int
    tags : list[str]
    ai_difficulty_rating : int
    human_difficulty_rating : str
    json_question : list[JSON_QUESTION_SCHEMA]
````

````python
def create_question_schema(q):
    try:
        return QUESTION_SCHEMA(**q)
    except ValidationError as e:
        print(f"Validation error for question {q.get('id', 'unknown')}:")
        print(e)
        return None


pp_data = past_papers_data['questions']
# Convert the data to QUESTION_SCHEMA objects
pp_data_objects = [create_question_schema(q) for q in pp_data]

# Remove any None values (questions that failed validation)
pp_data_objects = [q for q in pp_data_objects if q is not None]
````

````python
def generate_markdown(questions: list[QUESTION_SCHEMA]) -> str:
    markdown = ""
    for q in questions:
        # Generate a unique ID
        unique_id = str(uuid4())

        # Start the question block for the entire question set
        markdown += f"""---
~~~ 
id: "{unique_id}"
title: "Question {q.id}"
difficulty: {q.ai_difficulty_rating}
isSample: true
type: "QUESTION"
qaState: "PENDING"
~~~
"""

        for idx, sub_q in enumerate(q.json_question, start=1):
            # Add each sub-question under the main question title
            markdown += f"""
:::q{{marks={q.total_marks}}}
{sub_q.question}
:::

"""

            if sub_q.type == "options":
                markdown += ":::ain{choice}\n"
                for choice in sub_q.choices:
                    is_correct = "- [X]" if choice == sub_q.answer else "- [ ]"
                    markdown += f"{is_correct} {choice}\n"
                markdown += ":::\n\n"
            
            elif sub_q.type == "free-latex":
                markdown += f""":::ain{{free-latex}}
- {sub_q.answer}
:::

"""

            # Add the note if diagram_description is not empty
            if sub_q.diagram_description.strip():
                markdown += f""":::callout{{importantNote}}
{sub_q.diagram_description}
:::

"""

            # Always add the qs tag for each sub-question
            markdown += f""":::qs
{sub_q.answer}
:::\n
"""

    return markdown
````

````python
# Generate the markdown
markdown_content = ((generate_markdown(pp_data_objects)).replace('\t' , '\\t')).replace('\\\t' , '\\t')
````

````python
# Write to a file
with open("./worksheets/worksheet_1.txt", "w", encoding="utf-8") as file:
    file.write(markdown_content)
print("Questions have been written to questions.txt")
````
